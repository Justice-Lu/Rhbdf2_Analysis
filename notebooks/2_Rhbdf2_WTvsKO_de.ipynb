```python
import pandas as pd 
import numpy as np 
import os
import plotly.graph_objects as go
```

```python
Blobel_df = pd.read_csv('../DE_out/DE_allgene_WTvsiRhom2.csv', index_col=0)
Rhbdf2_df = pd.read_csv('../output/Rhbdf2_expression.csv', index_col=0)[['Rhbdf2_counts','Olfr']]
```

```python
# group the original dataframe by 'Group' and get the average count and sum
grouped = Rhbdf2_df.groupby('Olfr')['Rhbdf2_counts'].agg(['mean', 'sum']).reset_index()

# create a new dataframe with the average count and sum for each group
Rhbdf2_df = pd.DataFrame({'Olfr': grouped['Olfr'],
                       'Rhbdf2_avg': grouped['mean'],
                       'Rhbdf2_sum': grouped['sum']})

# merge the original dataframe with the new dataframe
# merged_df = pd.merge(df, new_df, on='Group')
```

```python
top_Rhbdf2_avg_Olfr = list(Rhbdf2_df.sort_values('Rhbdf2_avg', ascending=False)\
                    .head(int(len(Rhbdf2_df) * 0.2))['Olfr'])
bottom_Rhbdf2_avg_Olfr = list(Rhbdf2_df.sort_values('Rhbdf2_avg', ascending=True)\
                    .head(int(len(Rhbdf2_df) * 0.8))['Olfr'])
top_Rhbdf2_sum_Olfr = list(Rhbdf2_df.sort_values('Rhbdf2_sum', ascending=False)\
                    .head(int(len(Rhbdf2_df) * 0.2))['Olfr'])
bottom_Rhbdf2_sum_Olfr = list(Rhbdf2_df.sort_values('Rhbdf2_sum', ascending=True)\
                    .head(int(len(Rhbdf2_df) * 0.8))['Olfr'])
```

```python
"""
Volcano plot
"""

fig = go.Figure()
# plot padj > 0.05
# temp = Blobel_df
temp = Blobel_df.query('FDR > 0.05')
fig.add_trace(go.Scatter(x=temp['logFC'], 
                        y=-np.log10(temp['FDR']),
                        text=temp['symbol'],
                        mode='markers', 
                        name = 'FDR > 0.05',
                        marker=dict(size = 10, color = 'grey', opacity=0.3)))
temp = Blobel_df.query('FDR <= 0.05')
fig.add_trace(go.Scatter(x=temp['logFC'], 
                        y=-np.log10(temp['FDR']),
                        text=temp['symbol'],
                        mode='markers', 
                        name = 'FDR <= 0.05',
                        marker=dict(size = 10, color = 'red', opacity=0.3)))

# temp = Blobel_df[Blobel_df['symbol'].fillna('').str.contains('^Olfr')]
# temp = Blobel_df[Blobel_df['symbol'].isin(top_Rhbdf2_avg_Olfr)]
# # temp = Blobel_df[Blobel_df['symbol'].isin(top_Rhbdf2_sum_Olfr)]
# fig.add_trace(go.Scatter(x=temp['logFC'], 
#                         y=-np.log10(temp['FDR']),
#                         text=temp['symbol'],
#                         mode='markers', 
#                         name='Top Rhbdf2 avg Olfr',
#                         marker=dict(size = 10, color = 'red', opacity=0.3)))

# temp = Blobel_df[Blobel_df['symbol'].isin(bottom_Rhbdf2_avg_Olfr)]
# # temp = Blobel_df[Blobel_df['symbol'].isin(bottom_Rhbdf2_sum_Olfr)]
# fig.add_trace(go.Scatter(x=temp['logFC'], 
#                         y=-np.log10(temp['FDR']),
#                         text=temp['symbol'],
#                         mode='markers', 
#                         name='Bottom Rhbdf2 avg Olfr',
#                         marker=dict(size = 10, color = 'blue', opacity=0.3)))

# Add the horizontal line at y=0.5
fig.add_shape(type='line', x0=-10, x1=10,
                      y0=-np.log10(0.05), y1=-np.log10(0.05),
              line=dict(color='violet', width=3, dash='dash'))

fig.update_traces( 
    textposition='top center',
    hovertemplate =
    '<b>%{text}</b>' + 
    '<br>LogFC: %{x}'+
    '<br>FDR: %{y}<br>')

fig.update_layout(
    title='Rhbdf2 DE',
    autosize=True,
#     width=500,
#     height=500,
    template='simple_white'
)
fig.show()
fig.write_html("../output/vol_plot/WTvsRhbdf2.html")
```
